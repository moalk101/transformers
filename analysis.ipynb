```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(r"C:\Users\modar\Desktop\Uni\transforemer\models\Transformermodel\v4\log\loss_train.csv")
df_trimmed = df.iloc[5:]

time_steps = df_trimmed['Step']

plt.figure(figsize=(10, 6))
plt.plot(time_steps, df_trimmed['Value'], linestyle='-', linewidth=2)
plt.xlabel('Time Step')
plt.ylabel('Loss')
plt.title('Loss over Time Steps')
plt.grid(True, alpha=0.3)
plt.savefig("trainloss_plot.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
